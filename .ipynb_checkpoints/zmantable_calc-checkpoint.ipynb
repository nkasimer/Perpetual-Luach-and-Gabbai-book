```python
import math
def cos_mult(a,b):
    return math.cos(math.radians(a))*math.cos(math.radians(b))
def tan_mult(a,b):
    return math.tan(math.radians(a))*math.tan(math.radians(b))
```

```python
import datetime
```

```python
def costan_to_str(number):
    number = round(number*100)
    if number == 0:
        return " & 0"
    elif number == 100:
        return " & 100"
    elif number < 10:
        return " & 0"+str(number)
    else:
        return "& "+str(number)
```

```python
def sin_div_cos(angle,number):
    result = math.sin(math.radians(angle))/number
    result_display = round(result*100)
    if result_display > 200:
        return None
    else:
        return result_display
```

```python
def make_lat_line(lat):
    line = r"\multirow{2}{*}{"+str(lat)+r"°}&\textit{cos}"#×cos}"
    for dec in range(24):
        result = cos_mult(dec,lat)
        line += costan_to_str(result)
    line += r"\\* \space&\textit{tan}"#"×tan}"
    for dec in range(24):
        result = tan_mult(dec,lat)
        line += costan_to_str(result)
    line += r"\\\hline"
    return line
```

```python
def make_table1():
    rows = []
    definition = r"\begin{tiny}\begin{longtable}{c "
    for i in range(24):
        definition += "c |"
    definition += "c}"
    rows.append(definition)
    header = r"Lat&Dec"
    for i in range(24):
        header += "&"+str(i)+"°"
    header += r"\\\hline"
    rows.append(header)
    for lat in range(71):
        if lat in [66,67,68,69]:
            continue
        rows.append(make_lat_line(lat))
    rows.append(r"\end{longtable}\end{tiny}")
    return rows
```

```python
rows = make_table1()
for row in rows:
    print(row)
```

```python
useful_angles = [.833,5.5,6,6.5,7.1,8.5,10.2,11,11.5,14,16.1,18,19.75]
```

```python
def make_angle_line(number):
    result = str(round(number*100))
    for angle in useful_angles:
        result += " & "
        value = sin_div_cos(angle,number)
        if value != None:
            result += str(value)
        else:
            result += "-"
    result += r"\\"
    return result
```

```python
def make_table2():
    rows = []
    definition = r"\begin{scriptsize}\begin{longtable}{"
    for i in range(14):
        definition += "c |"
    definition += "c}"
    rows.append(definition)
    header = r"\space "
    for angle in useful_angles:
        header += "&"+str(angle)+"°"
    header += r"\\\hline"
    rows.append(header)
    i = 0.01
    while i <= 1.001:
        rows.append(make_angle_line(i))
        i += 0.01
    rows.append(r"\end{longtable}\end{scriptsize}")
    return rows
```

```python
rows = make_table2()
for row in rows:
    print(row)
```

```python
def make_arcsin_table():
    rows = [r"\begin{footnotesize}\begin{longtable}{c | c}"]
    rows.append(r"Previous Result:&Minutes\\\hline")
    for i in range(101):
        value = i/100
        degrees = math.degrees(math.asin(value))
        minutes = degrees * 4
        rows.append(str(i)+"&"+str(round(minutes))+r"\\\hline")
    rows.append(r"\end{longtable}\end{footnotesize}")
    return rows
```

```python
rows = make_arcsin_table()
for row in rows:
    print(row)
```

```python
def eqn_time(day):
    B = math.radians((day-1)*360/365)
    return -229.2*(0.000075+0.001868*math.cos(B)-0.032077*math.sin(B)-0.014615*math.cos(2*B)-0.04089*math.sin(2*B))
```

```python
def dec(day):
    return -23.44*math.cos(2*math.pi*(day+10)/365)
```

```python
def date_table():
    rows = [r"\begin{longtable}{c | c | c | c}"]
    rows.append(r"Date&Day Number&Equation of Time&Declination\\\hline")
    start_date = datetime.datetime(2023,1,1)
    end_date = datetime.datetime(2024,1,1)
    eachdate = start_date
    while eachdate<end_date:
        line = calc_date(eachdate)
        rows.append(line)
        eachdate = eachdate + datetime.timedelta(days=3)
    rows.append(r"\end{longtable}")
    return rows
```

```python
def calc_date(date):
    day_num = int(date.strftime("%j"))
    time = eqn_time(day_num)
    declin = dec(day_num)
    day = date.strftime("%b-%d")
    return day + " & "+str(day_num)+" & "+str(round(time))+" & "+str(round(declin))+r"\\\hline"
```

```python
rows = date_table()
for row in rows:
    print(row)
```
