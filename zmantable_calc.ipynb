```python
import math
def cos_mult(a,b):
    return math.cos(math.radians(a))*math.cos(math.radians(b))
def tan_mult(a,b):
    return math.tan(math.radians(a))*math.tan(math.radians(b))
```

```python
import datetime
```

```python
def costan_to_str(number):
    number = round(number*100)
    if number == 0:
        return " & 0"
    elif number == 100:
        return " & 100"
    elif number < 10:
        return " & 0"+str(number)
    else:
        return "& "+str(number)
```

```python
def sin_div_cos(angle,number):
    result = math.sin(math.radians(angle))/number
    result_display = round(result*100)
    if result_display > 200:
        return None
    else:
        return result_display
```

```python
def make_lat_line(lat):
    line = r"\multirow{2}{*}{"+str(lat)+r"°}&\textit{cos}"#×cos}"
    for dec in range(24):
        result = cos_mult(dec,lat)
        line += costan_to_str(result)
    line += r"\\* \space&\textit{tan}"#"×tan}"
    for dec in range(24):
        result = tan_mult(dec,lat)
        line += costan_to_str(result)
    line += r"\\\hline"
    return line
```

```python
def calc_mins(dec,lat,deg):
    tan_term = math.tan(math.radians(dec))*math.tan(math.radians(lat))
    cos_term = math.sin(math.radians(deg))/(math.cos(math.radians(dec))*math.cos(math.radians(lat)))
    try:
        return 4*math.degrees(math.asin(tan_term + cos_term))
    except:
        return None
```

```python
def eqn_time(day):
    B = math.radians((day-1)*360/365)
    return -229.2*(0.000075+0.001868*math.cos(B)-0.032077*math.sin(B)-0.014615*math.cos(2*B)-0.04089*math.sin(2*B))
```

```python
def dec(day):
    return -23.44*math.cos(2*math.pi*(day+10)/365)
```

```python
def date_table(precise):
    rows = [r"\begin{longtable}{c | c | c}"]
    rows.append(r"Date&Equation of Time&Declination\\\hline")
    start_date = datetime.datetime(2023,1,1)
    end_date = datetime.datetime(2024,1,1)
    eachdate = start_date
    while eachdate<end_date:
        line = calc_date(eachdate,precise)
        rows.append(line)
        if precise == True:
            eachdate = eachdate + datetime.timedelta(days=1)
        else:
            eachdate = eachdate + datetime.timedelta(days=3)
    rows.append(r"\end{longtable}")
    return rows
```

```python
def calc_date(date,precise):
    day_num = int(date.strftime("%j"))
    time = eqn_time(day_num)
    declin = dec(day_num)
    day = date.strftime("%b-%d")
    if precise == True:
        minutes = math.floor(time)
        seconds = round(60*(time - minutes))
        if seconds<10:
            seconds = "0"+str(seconds)
        declin_degrees = math.floor(declin)
        declin_mins = round(60*(declin - declin_degrees))
        return day + " & "+str(minutes)+":"+str(seconds)+" & "+str(declin_degrees)+"°"+str(declin_mins)+r"'\\\hline"
    else:
        return day + " & "+str(round(time))+" & "+str(round(declin))+r"°\\\hline"
```

```python
def mins_minssecs(time):
    result = [0,0]
    result[0] = math.floor(time)
    result[1] = round(60*(time - result[0]))
    return result
```

```python
def mins_per_degree_calc(dec,lat):
    return 4 * math.degrees(math.radians(1)/cos_mult(dec,lat))
```

```python
def time_to_str(number):
    if number<10:
        return "0"+str(number)
    elif number == 0:
        return "00"
    else:
        return str(number)
```

```python
def linear_approx_line(lat):
    line = str(lat)+"°"
    for dec in range(24):
        line = r"\multirow{2}{*}{"+str(lat)+r"°}"#×cos}"
    for dec in range(24):
        result = 4*math.degrees(tan_mult(dec,lat))
        time = mins_minssecs(result)
        if time[0]>=60:
            hours = math.floor(time[0]/60)
            minutes = time[0]%60
            mins_str = time_to_str(minutes)
            seconds = time[1]
            secs_str = time_to_str(seconds)
            line += "&"+str(hours)+":"+mins_str+":"+secs_str
        else:
            line += "&"+str(time[0])+":"+time_to_str(time[1])
    line += r"&\multirow{2}{*}{"+str(lat)+r"°}\\* \space" 
    for dec in range(24):
        result = mins_per_degree_calc(dec,lat)
        time = mins_minssecs(result)
        secs = time[1]
        secs_str = time_to_str(secs)
        line += "&"+str(time[0])+":"+secs_str
    line += r"&\space\\\hline"
    return line
```

```python
def sin_error(angle):
    result_decimal = math.degrees(math.sin(math.radians(angle)))
    result_degrees = math.floor(result_decimal)
    result_minutes = round(60*(result_decimal - result_degrees))
    if result_minutes == 60:
        result_degrees += 1
        result_minutes = 0
    return str(result_degrees)+"°"+str(result_minutes)+r"'"
```

```python
def mult_table():
    rows = [r"\begin{footnotesize}\begin{tabular}{c | c | c | c | c | c | c | c | c | c | c | c | c | c | c}"]
    first_row = "\space"
    corrected_angles = []
    for angle in useful_angles:
        angle_deg = math.floor(angle)
        angle_mins = round(60*(angle-angle_deg))
        first_row += "&"+str(angle_deg)+"°"
        if angle_mins != 0:
            first_row += str(angle_mins)+r"'"
        corrected_angles.append(math.degrees(math.sin(math.radians(angle))))
    first_row += r"\\\hline"
    rows.append(first_row)
    for i in range(0,49):
        row_time = 4+i/6
        row_mins = math.floor(row_time)
        row_secs = round((row_time - row_mins)*60)
        if row_secs == 0:
            row = str(row_mins)+":00"
        else:
            row = str(row_mins)+":"+str(row_secs)
        for angle in corrected_angles:
            time = angle*row_time
            minutes = math.floor(time)
            seconds = round(60*(time - minutes))
            if seconds == 60:
                minutes += 1
                seconds_str = "00"
            elif seconds == 0:
                seconds_str = "00"
            elif seconds < 10:
                seconds_str = "0"+str(seconds)
            else:
                seconds_str = str(seconds)
            if minutes >= 60:
                hours = math.floor(minutes/60)
                minutes = minutes%60
                if minutes==0:
                    minutes_str = "00"
                elif minutes < 10:
                    minutes_str = "0"+str(minutes)
                else:
                    minutes_str = str(minutes)
                row += "&"+str(hours)+":"+minutes_str+":"+seconds_str
            else:
                row += "&"+str(minutes)+":"+seconds_str
        if i != 61:
            row += r"\\\hline"
        rows.append(row)
    rows.append(r"\end{tabular}\end{footnotesize}")
    return rows
```
