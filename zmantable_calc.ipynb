```python
import math
def cos_mult(a,b):
    return math.cos(math.radians(a))*math.cos(math.radians(b))
def tan_mult(a,b):
    return math.tan(math.radians(a))*math.tan(math.radians(b))
```

```python
def costan_to_str(number):
    number = round(number*1000)
    if 0<number<10:
        return " & 00"+str(number)
    elif number < 100:
        return " & 0"+str(number)
    else:
        return "& "+str(number)
```

```python
costan_to_str(.02)
```

```python
def make_lat_line(lat):
    line = r"\multirow{2}{*}{"+str(lat)+r"°}&\textit{cos×cos}"
    for dec in range(24):
        result = cos_mult(dec,lat)
        line += costan_to_str(result)
    line += r"\\* \space&\textit{tan×tan}"
    for dec in range(24):
        result = tan_mult(dec,lat)
        line += costan_to_str(result)
    line += r"\\\hline"
    return line
```

```python
def make_table():
    rows = []
    definition = r"\begin{longtable}{c "
    for i in range(23):
        definition += "c |"
    definition += "c}"
    rows.append(definition)
    header = r"\backslashbox{Latitude}{Declination}"
    for i in range(24):
        header += "&"+str(i)+"°"
    header += r"\\"
    rows.append(header)
    for lat in range(70):
        rows.append(make_lat_line(lat))
    rows.append(r"\end{longtable}")
    return rows
```

```python
rows = make_table()
for row in rows:
    print(row)
```
